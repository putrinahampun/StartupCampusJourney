```python
# DATAFRAME MERGING => menggabungkan dua atau lebih dataframe.

# leftjoin
import pandas as pd

dict_df1 = {'id':[1,2,3,4,5],
            'movie':['titanic','kimi no nawa','doraemon','lady bird','larva']}

dict_df2 = {'id':[1,2,3,4,5,6,7,8],
            'actor':['hermonie','harry','suzuke','nobita','jerome','otoke','sibuyaa','nana'],
            'hobbies':['swim','read','sing','dance','sleep','eat','talk','jump']}

df1 = pd.DataFrame(dict_df1)
df2 = pd.DataFrame(dict_df2)

df_merge = df1.merge(df2, how='left', left_on='id',right_on='id')
df_merge
```

```python
# right join
df_mergee = df1.merge(df2, how='right',left_on = 'id', right_on='id')
df_mergee
```

```python
# Inner join
df_merge = df1.merge(df2, how='inner', left_on='id',right_on='id')
df_merge
```

```python
# outer join
df_merge = df1.merge(df2, how='outer', left_on='id', right_on='id')
df_merge
```

```python
# cross join
import pandas as pd

dict_df1 = {'id':[1,2,3]}
dict_df2 = {'name':['nana','echan','jeno']}

df1 = pd.DataFrame(dict_df1)
df2 = pd.DataFrame(dict_df2)

df1['key'] = 1
df2['key'] = 1

df_merge = pd.merge(df1,df2,on='key').drop('key',1)
df_merge
```

## HANDLING MISSING VALUES

1.   jumlah data yang hilang lebih dari 60% => dihapus
2.   mengisi kategori yang hilang dengan kategori paling banyak (other)
3.   numeric jika hilang boleh diisi dengan nilai rata2, median, atau 0.
4.   fungsi2 yang digunakan, yaitu:
    * dropna() => membuang nilai NaN
    * fillna() => mengisi nilai kosong menjadi NaN.



```python
# dropna dan fillna
import pandas as pd
import numpy as np

data_numerik = [2,11,np.nan,15,8,3,9,12,6,7,np.nan,np.nan]
data_kategorik = ['kecil','besar','kecil','besar','kecil','kecil','besar',np.nan,np.nan,'kecil','besar',np.nan]
data_null = [np.nan,np.nan,np.nan,np.nan,np.nan,'isi','isi',np.nan,np.nan,'isi','isi',np.nan]
data_tinggi = ['tinggi','tinggi','tinggi','tinggi','rendah','rendah',np.nan,np.nan,'rendah','tinggi','rendah','rendah']

df = pd.DataFrame(list(zip(data_numerik, data_kategorik,data_null,data_tinggi)),
                  columns = ['number','category','null_data','tinggi'])

# kolom category akan kita isi dengan besar
df['category'] = df['category'].fillna('besar')

# kolom number akan diisi angka 10
df['number'] = df['number'].fillna(10)

df = df.dropna()
df
```

```python
# One Hot Encoding
import pandas as pd

data = pd.DataFrame({
    'color':['red','blue','green','blue']
})

df_encoded = pd.get_dummies(data)
df_encoded
```

```python
# One Hot Encoding
import pandas as pd

data = pd.DataFrame({
    'color':['red','blue','green','blue']
})

df_encoded = pd.get_dummies(data,drop_first=True)
df_encoded
```

```python
# label encoding
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.DataFrame({
    'color':['red','blue','green','blue']
})

encoder = LabelEncoder()
encoded_data = encoder.fit_transform(data['color'])

data['Color_Encoded'] = encoded_data

data
```
