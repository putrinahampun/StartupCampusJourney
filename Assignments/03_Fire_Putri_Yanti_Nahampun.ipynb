# Pandas Question

1. What is the minimum, median, max, and average salary for employee by Married Description and Gender.
Hints: use agg function<br>
Expected Output as below<br>
(5 Points)

```python
# 1. What is the minimum, median, max, and average salary for employee by Married Description and Gender
# Hints: use agg function
# Expected Output as below
# (5 Points)

import pandas as pd

# SELECT THE MARITAL DESC, SEX, AND SALARY ATRIBUTES
df = data[['MaritalDesc','Sex','Salary']]

# USING AGGREGATE FUNCTION TO COUNT MIN, MEDIAN, MAX, AND AVERAGE OF SALARIES.
df.groupby(
    ['MaritalDesc','Sex']).agg(
        min_salary=('Salary','min'),
        median_salary=('Salary','median'),
        max_salary=('Salary','max'),
        average_salary=('Salary','mean')
    )
```

2. What is top-5 reason of termination?<br>
Expected Output as below<br>
(5 Points)

```python
#code here
import pandas as pd

# SELECT TERMINATION_REASON ATRIBUTE
tR = data[['TermReason']]

# USING AGGREGATION FUNCTION TO COUNT TOTAL TERMINATION EACH TERM REASON.
tR.groupby(
    ['TermReason']).agg(
        total_termination = ('TermReason','count')).sort_values(
            # using sort function for sorting the total from the highest score to the lowest
            by = ['total_termination'], ascending = False
        # using head function to print top 5 data.
        ).head(5)
```

3. What is the highest number of employee by Recruitment Source that 'Exceeds' the Performance Score?

Expected Output as below<br>
(5 Points)

```python
#code here
import pandas as pd

# SELECT TERMINATION REASON AND PERFORMANCE SCORE (EQUALS WITH EXCEEDS) ATRIBUTES.
tR = data.loc[(data.PerformanceScore == 'Exceeds'),['RecruitmentSource','PerformanceScore']]

# USING AGGREGATION FUNCTION TO COUNT TOTAL RECURITMENT WITH EXCEDDS STATUS EACH RECRUITMENT SOURCE.
tR.groupby(
    ['RecruitmentSource']).agg(
        total_recruit = ('PerformanceScore','count')).sort_values(
            # using sort function for sorting the total from the highest score to the lowest
            by = ['total_recruit'], ascending = False
        )
```

7. Buatlah visualisasi barchart untuk menghitung jumlah "termination of an employee's" berdasarkan department.<br>
(5 points)<br>
*bentuk output di bawah hanya contoh*

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# USE AGG FUNCTION TO COUNT TOTAL OF PEOPLE TERMINATE BASED ON EACH DEPARTMENT
d = data.groupby(['Department']).agg(
    total_termination = ('Termd','sum')).sort_values(
        by=['total_termination'], ascending=True).reset_index()

# DEFINE THE COLOR OF BAR
c = ['yellow','orange','purple','green','blue','red']

# SETTING FOR THE GRAPH
fig, ax = plt.subplots(figsize=(18, 6))

# CALCULATE THE POSITION OF THE GRAPH
positions = np.arange(len(d))

# DEFINE THE WIDTH FOR EACH GRAPH
bar_width = 0.9

# ITERATE THE DATA AND CREATE BARS FOR EACH DEPARTMENT
for i, row in d.iterrows():
    ax.barh(positions[i], row['total_termination'], color=c[i], height=bar_width, label=row['Department'])

# SET Y LABELS (DEPARTMENT)
ax.set_yticks(positions)
ax.set_yticklabels(d['Department'])

# CUSTOMIZE LABELS
ax.set_xlabel('Termd')
ax.set_ylabel('Department')
ax.legend(title='Department')

plt.show()
```

# Create Machine Learning Model to Predict Termd and Question

```python
#11. Buat fitur baru (kolom baru) berupa "Age" dalam satuan tahun. Untuk mendapatkannya, ambil data pada kolom DOB dikurangi dengan 2022
#(5 points)

import pandas as pd
import datetime as dt

# CONVERT 'DOB' COLUMN TO DATETIME FORMAT
data['DOB_date'] = pd.to_datetime(data['DOB'], format='%m/%d/%y')

# CALCULATE THE CURRENT DATE
current_date = dt.datetime(2022, 12, 12)  # Assuming December, 2022 as the reference date

# CALCULATE THE AGE BY SUBSTRACTING THE CURRENT DATE WITH THE DOB DATE
data['Age'] = (current_date - data['DOB_date']).astype('<m8[Y]')

# ROUNDING THE AGE VALUE
data['Age'] = data['Age'].apply(round)

data[['DOB','DOB_date','Age']]
```

```python
#12. Lakukan MinMaxScaler dari fitur "Salary"
# (5 points)

from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# INITIALIZE THE MINMAXSCALER()
scaler = MinMaxScaler()

scaler.fit(data[['Salary']])
data['Salary_minmax'] = scaler.transform(data[['Salary']])

data[['Salary','Salary_minmax']]
```

```python
#13. Ubahlah semua tipe data object menjadi category
# 5 point

# BEFORE THE CHANGE
data.dtypes
```

```python
# AFTER THE CHANGE
for col in data.columns:
  if data[col].dtypes == 'object':
    data[col] = data[col].astype('category')

data.dtypes
```

```python
#14. Ubahlah "RecruitmentSource" dengan menggunakana Label Encoder
# 5 point

from sklearn.preprocessing import LabelEncoder

# INITIALIZE THE LABEL ENCODER
label_encoder = LabelEncoder()
data['RecruitmentSource'] = label_encoder.fit_transform(data['RecruitmentSource'])

#15. Ubahlah "PerformanceScore" dengan menggunakana Label Encoder
# 5 point
data['PerformanceScore'] = label_encoder.fit_transform(data['PerformanceScore'])
```

```python
#16. Pisahkan dependent variable dan independent variable
# 5 point
#Dependent Varible (target)
y = data['Termd']
y.head()

#Independent Variable (feature)
x = data[['MarriedID','MaritalStatusID',	'GenderID','DeptID','PerfScoreID','PositionID','Age',
          'RecruitmentSource','PerformanceScore','Salary_minmax','Absences','SpecialProjectsCount',
          'EmpSatisfaction','EngagementSurvey']]
x.head()
```

```python
#17. Bagi data menjadi Train-Test sebesar 80:20
# (5 points)

from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score, precision_recall_curve, roc_curve, precision_score, recall_score

x_train, x_test, y_train, y_test = tts(x,y,test_size=.2,random_state=0)
```

```python
# 18. Lakukan pemodelan machine learning (metode klasifikasi bebas) untuk bisa memprediksi status "Termd". Lakukan evaluasi hasil dengan menampilkan
# score klasifikasi, precision dan recall.
# Gunakan model yang menurut kalian terbaik
# (5 points)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(tol=0.0001, C=0.01, max_iter=1000)

train_model = model.fit(x_train, y_train)

# make predictions for test data
y_pred = train_model.predict(x_test)
predictions = [round(value) for value in y_pred]

# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)
print("Accuracy  %.2f%%" % (accuracy * 100.0))
print("Precision : %.2f%%" % (precision* 100.0))
print("Recall : %.2f%%" % (recall * 100.0))
```
