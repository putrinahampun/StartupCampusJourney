```python
import pandas as pd
import numpy as np
```

```python
filename = 'income.csv'

# header = None => kalau data tidak punya header
data = pd.read_csv(filename, header=None)
```

```python
columns_name =['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship',
               'race','sex','capital_gain','capital_loss','hours_per_week','native_country','income']
```

```python
data.columns = columns_name
data.head()
```

### SPLIT INPUT (FEATURES) AND TARGET(LABEL)

```python
# Features / inputs
x = data.drop(['income'],axis=1)
x.head(3)
```

```python
type(x)
```

```python
# Target / Label
y = data['income']
y.head(3)
```

```python
# input nya bertipe data series
# kumpulan dari series membentuk dataframe
type(y)
```

### TRAIN AND TEST SPLIT

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train, y_test = tts(x,y,test_size=.3,random_state=0)
```

```python
print('Feature size for training:', x_train.shape)
print('Target size for training:', y_train.shape)
print('Feature size for testing:', x_test.shape)
print('Target size for testing:', y_test.shape)
```

```python
for i in [x_train, x_test]:
  i['workclass'].fillna(x_train['workclass'].mode()[0], inplace=True)
  i['occupation'].fillna(x_train['occupation'].mode()[0], inplace=True)
  i['native_country'].fillna('others', inplace=True)
```

```python
x_train.workclass.unique()
```

### ONE-HOT ENCODING

```python
import category_encoders as ce
```

```python
encoder = ce.OneHotEncoder(cols=categorical[:-1])
encoder
```

```python
x_train_encoder = encoder.fit_transform(x_train)
x_train_encoder.drop(['native_country_42'], axis=1, inplace=True)
x_train_encoder.head()
```

```python
x_test_encoder = encoder.fit_transform(x_test)
x_test_encoder.head()
```

```python
from sklearn.preprocessing import RobustScaler

# Feature Scaling
scaler = RobustScaler()
x_train_scaler = scaler.fit_transform(x_train_encoder)
x_test_scaler = scaler.fit_transform(x_test_encoder)
```

```python
x_train_scaler
```

```python
x_test_scaler
```

```python
x_train_scaler = pd.DataFrame(x_train_scaler, columns=x_train_encoder.columns)
x_train_scaler.head()
```

```python
x_test_scaler = pd.DataFrame(x_test_scaler, columns=x_test_encoder.columns)
x_test_scaler.head()
```

```python
x_test_scaler.isnull().any()
```

### MODELING

```python
from sklearn.ensemble import RandomForestClassifier as RF
```

```python
model = RF()
model.fit(x_train_scaler,y_train)
```

```python
y_pred = model.predict(x_test_scaler)
y_pred
```

```python
result = pd.DataFrame({
    'NILAI_ASLI':y_test,
    'NILAI_PREDIKSI':y_pred
})
```
